```python
# Load files
train = pd.read_csv('/content/sample_data/california_housing_train.csv')
test = pd.read_csv('/content/sample_data/california_housing_test.csv')
train.head()
```

#Modelling!

```python
#scale numerical features for logistic model
features = train.drop(columns=['median_house_value']).columns
target = 'median_house_value'
```

```python
#split train data into train and validation set
X_train, X_test, y_train, y_test = train_test_split(train[['longitude','latitude','population']],
                                                    train[target].to_frame(),
                                                    test_size=0.3,
                                                    random_state=SEED)
```

```python
#fit Lightgbm model on rrain data
model = LGBMRegressor()
model.fit(X_train, y_train['median_house_value'])

#make predictions on validation set
preds = model.predict(X_test)
```

```python
#check performance of the model
print(mean_absolute_error(y_test, preds))
```

```python
data={'model':model,'latitude':train['latitude'],'longitude': train['longitude'],'population':train['population']}
with open('saved_steps.pkl','wb') as file:
  pickle.dump(data,file)
```
